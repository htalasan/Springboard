## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## Find the top 10 major project themes


```python
# comment the data set that you don't want to use and obviously leave the one you want to use commented

world_bank_project_themes = world_bank_df['mjtheme_namecode']
#world_bank_project_themes = world_bank_df['theme_namecode'].dropna()

#world_bank_project_themes
```

```python
s = world_bank_project_themes
```

```python
#world_bank_df['theme_namecode']
```

```python
project_themes_list = []

for project in world_bank_project_themes.tolist():
    for d in project:
        project_themes_list.append(d)

```

```python
#project_themes_list
```

```python
# find out how many times a code of a theme appears
code_counts = {}

for project in project_themes_list:
    if project['code'] in code_counts.keys():
        code_counts[project['code']] += 1
    else:
        code_counts[project['code']] = 1

# sort and print top 10 codes of themes
sorted_code_counts = [(k, code_counts[k]) for k in sorted(code_counts, key=code_counts.get, reverse=True)]
sorted_code_counts[:10]
```

```python
# find out how many times a name of a theme appears

name_counts = {}

for project in project_themes_list:
    if project['name'] in name_counts.keys():
        name_counts[project['name']] += 1
    else:
        name_counts[project['name']] = 1

# sort and print top 10 names of themes
sorted_name_counts = [(k, name_counts[k]) for k in sorted(name_counts, key=name_counts.get, reverse=True) if k != '']
sorted_name_counts[:10]
```

## Create a dataframe with the missing names filled in

```python
# create a dictionary with the codes and names
project_themes_dict = {}

for theme in project_themes_list:
    if theme['name'] != '': # skip names that are blank
        project_themes_dict[theme['code']] = theme['name']
    
#project_themes_dict
```

```python
def fill_in_themename(l):
    '''Takes in the list of dictionaries of namecodes. If a name is missing, it will fill in the name based on the code.'''
    
    for d in l:
        if d['name'] == '':
            d['name'] = project_themes_dict[d['code']]
    
    return l
```

```python
world_bank_project_themes.apply(fill_in_themename)

# check to see that all names have been filled in

project_themes_list = [] 

for project in world_bank_project_themes.tolist():
    for d in project: # each element in the themes is a list of dictionaries
        project_themes_list.append(d)

name_counts = {}

for project in project_themes_list:
    if project['name'] in name_counts.keys():
        name_counts[project['name']] += 1
    else:
        name_counts[project['name']] = 1
        
#name_counts 
```

```python
# check to see if the same

[v for k, v in code_counts.items()] == [v for k, v in name_counts.items()]
```

```python
len(code_counts)
```

After applying the function, fill_in_themename to the 'mjtheme_namecode' column of the dataframe, we can see that the name_counts and the code_counts are the same.
